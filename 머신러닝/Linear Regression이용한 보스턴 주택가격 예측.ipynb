```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy import stats
from sklearn.datasets import load_boston 
```

```python
#boston 데이터셋 로드
boston = load_boston()
```

```python
# boston 데이터셋 DataFrame변환
bostonDF = pd.DataFrame(boston.data,columns= boston.feature_names)
```

```python
# boston dataset의 target array는 주택가격. PRICE 컬럼으로 DataFrame에 존재
bostonDF['PRICE'] = boston.target
```

```python
print(bostonDF.shape)
bostonDF.head()
```

```python
# 2개의 행과 4개의 열을 가진 subplot을 이용. axs는 4X2의 ax를 가짐
fig,axs = plt.subplots(figsize=(16,8), ncols=4, nrows=2)
lm_featrues = ['RM','ZN','INDUS','NOX','AGE','PTRATIO','LSTAT','RAD']
for i, feature in enumerate(lm_featrues):
    row = int(i/4)
    col = i%4
    # 시본의 regplot을 이용해 산점도와 선형 회귀 직선을 함께 표현
    sns.regplot(x=feature, y='PRICE', data=bostonDF, ax=axs[row][col])
```
