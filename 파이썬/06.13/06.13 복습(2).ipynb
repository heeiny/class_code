파생변수생성

파생변수 생성 이유? 
기존의 데이터를 가지고 연산학습(경우의 수 늘리기 위함)
데이터프레임명["추가할 칼럼의 이름"] = 연산식 
shift의 행이 뭐지?? coloumn을 하나 내리는 함수 
shift(n)이 있다면,n개 만큼 column을 내림

fillna() -> ()안의 값을 결측치에 채워준다
fillna -> method 속성(매개변수)
method ->ffill, bfill (앞,뒤 행의 값을 채워준다)

diff() ->값의 차이를 표시, 행과 행, 열과 열의 차이를 출력

```python
 

corona["일일확진자"] = corona["decideCnt"] - corona["decideCnt"].shift(0)
## shift() : ()안에 아무것도 없으면 그 전의 행의 차를 그대로 대입함
## shift().fillna(0)을 사용하면 데이터 값이 채워지지 않음
```

```python
corona["일일사망자"] = corona["deathCnt"].diff().fillna(0)
corona.head(2)
```
