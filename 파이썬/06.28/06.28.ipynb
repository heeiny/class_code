```python
# 결측치 확인
corona_df.isna().sum()
```

```python
# 새로운 파생변수 생성
# 'accExamCnt'의 차이만큼을 일일확진자 컬럼에 추가. 
# 같은 컬럼의 값의 차이를 넣어주는 방법 2가지 
# shift(), diff()
# shift(n): n값만큼의 인덱스를 이동하여 그 행의 값을 가지고 오는 부분
# diff(): 한 객채 내의 열과 열/ 행과 행의 차이를 출력

corona_df["일일확진자"] = (corona_df["decideCnt"] - corona_df["decideCnt"].shift())
# 첫행의 값은 NaN 결측치로 출력이 된다.
# 이유는 전의 행의 값이 존재하지 않아서 연산이 불가능하기 때문
# 연산 후 결측치의 값을 채워주는 함수
# fillna()

```

```python
corona_df["일일확진자"] = (corona_df["decideCnt"] - corona_df["decideCnt"].shift()).fillna(0)
corona_df.head(2)
```

```python
corona_df["일일확진자"] = (corona_df["decideCnt"].diff().fillna(0))
corona_df["일일사망자"] = (corona_df["deathCnt"].diff().fillna(0))
corona_df.head(2)
```

# fillna() 
# ()안의 숫자형태나 문자형태의 데이터를 넣으면 결측치인 부분이 삽입 
# method = ('ffill', 'bfill')
# ffill : 전의 행의 값이 존재하면 그 값으로 결측치를 채워줌
# bfill : 후의 행의 값이 존재하면 그 값으로 결측치를 채워줌

```python
corona_df.head(10)
```

```python
# corona데이터는 누적의심자,누적확진률 컬럼의 결측치가 존재
# tail 확인했을 때 마지막 값이 존재하지 않는다.
# ffill을 사용해서 전의 행의 값을 가지고와서 결측치를 채워보는 예제
corona_df["accExamCnt"].fillna(method="ffill").tail(10)
```

```python
# apply함수 -> 컬럼에 있는 각 원소들에 연산을 해주는 함수(**여기 부분 낯섬)
# 컬럼에 있는 각 원소들에 연산을 해주는 함수
# lambda함수
# 표현식을 한줄로 처리하기 위해 사용하는 함수

# 외부함수를 이용하여 apply()를 사용할 때
# 데이터프레임[컬럼].apply(외부함수명): 인자값을 넣지 않는 것 -> 이유. apply를 통해 각 인자들이 한번씩 출력될 거니까!
# apply, lambda함수를 같이 사용하는 경우
# 데이터프레임[컬럼].apply(lambda x : 표현식)

# 누적확진률을 기준으로 1.55보다 높으면 'High'. 출력 낮으면 'Low'값을 가지는 파생변수 'H/L'

# 외부 함수를 이용하여 apply( )
def H_and_L(x):
    if x > 1.55:
        return "High"
    else:
        return "Low"
corona_df["H/L"] = corona_df["accDefRate"].apply(H_and_L)
```

```python
# apply lambda함수를 같이 사용하는 경우
corona_df["H/L"] = corona_df["accDefRate"].apply(lambda x: "High" if (x > 1.55) else "Low")
corona_df["H/L"]
```

```python
# 컬럼의 값들의 빈도수를 체크하는 함수
# value_counts()
corona_df["H/L"].value_counts()
```

## Concat() 
단순 행 추가하는  함수
axis = 0이면 행, 1이면 열 추가 (기본값은 행을 추가)

```python
transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction.head()
```

```python
transaction_detail = pd.concat([transaction_detail_1, transaction_detail_2], ignore_index=True)
```

Merge()
조인 결합을 이용해서 두 데이터 프레임을 합치는 함수

```python

# 특정한 조건을 기준으로 행을 추가해주는 함수 
# on 속성은 합쳐질 데이터프레임에 조건을 넣어주는 부분.
# 컬럼의 값들이 같은 경우 열을 추가해 줌
# on 속성에는 컬럼명을 적어주는 함수
# how 속성에는 기준이 되는 데이터프레임을 지정
# left,right,inner,outer 값들은 지정할 수 있음
join_data = pd.merge(transaction, transaction_detail, on = 'transaction_id', how='left' )
# transaction과 transaction_detail 조인 결합 기준은 left 조건은 transaction_id 값이 같은 경우
join_data.head()
```

### uriage 데이터 로드
1. item_name 데이터가 대/소문자 문제, 공백 문제 ->  그룹화 X -> 규칙성이 있게 데이터 변환
2. 결측치가 존재 -> 결측치의 값을 채워준다.
3. 상품별로 그룹화 확인

```python
# 데이터 로드
uriage_df = pd.read_csv("../csv/uriage.csv")
uriage_df.head(3)
```

```python
## item_name 데이터를 대문자로 변환
## case1
uriage_df["item_name"] = uriage_df["item_name"].str.upper()
## case2
uriage_df["item_name"] = uriage_df["item_name"].apply(lambda x : x.upper())
```

```python
## item_name 공백을 삭제
#case1
uriage_df["item_name"] = uriage_df["item_name"].str.replace(" ","")
#case2
uriage_df["item_name"] = uriage_df["item_name"].apply(lambda x : x.replace(" ",""))

## strip() -> 문자열 양 옆의 공백을 삭제
## replace(" ","") -> 공백을 삭제(문자열 사이의 공백도 포함)
## 두 함수의 차이는?
```

```python
uriage_df.isna().sum()
```

```python
# item_price 결측치에 item_name으로 그룹화 item_price의 평균 값을 삽입
flg_is_null = uriage_df["item_price"].isna()
flg_is_null
```

```python
# item_pirce의 값이 결측치인 경우 item_name의 유니크 값(중복 데이터 삭제)
na_list=(uriage_df.loc[flg_is_null, "item_name"].unique())              ##.uniqe는 중복값 삭제
na_list
```

```python
# '~' 연산자 -> bool의 형태를 반전
for i in na_list:
    ## (~flg_is_null)&(uriage_df["item_name"] == i): 결측치가 아니고 item_name의 값이 na_list 값 중 하나
    price = uriage_df.loc[(~flg_is_null)&(uriage_df["item_name"] == i), 
                            "item_price"].mean()            ## item_price 컬럼을 선택해서 그 값의 평균  
    ## price 값은 na_list중 하나의 값의 평균 값
    ## (flg_is_null)&(uriage_df["item_name"] == i): 결측치고, item_name의 값이 na_list 값 중 하나
    uriage_df.loc[(flg_is_null)&(uriage_df["item_name"] == i),
                                 "item_price"] = price      ## item_price에 price라는 변수를 삽입
```
