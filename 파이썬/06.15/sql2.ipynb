## job_code 새로운 파생변수 생성


#1. salesman -> 1 , manager -> 2 그 외의 직업은 3
#2. 사원번호, 이름, 직업, job_code 출력

```python
#case1  if문

def change(x):
    if x == "SALESMAN":
            return "1"
    elif x == "MANAGER":
            return "2"
    else:
            return "3"

```

```python
df["JOB_CODE"] = df["JOB"].apply(change)
```

```python
df.loc[:,["EMPNO","ENAME","JOB","JOB_CODE"]]
```

```python
#case2  apply. lambda함수 사용(pandas)

df["JOB_CODE"]=df["JOB"].apply(lambda x: "1" if (x == "SALESMAN") 
else ("2" if ( x == "MANAGER") else "3"))
df.loc[:,["EMPNO","ENAME","JOB","JOB_CODE"]]

    
```

```python
#case3 CASE WHEN END구문 

sql = """
SELECT EPNO,ENAME,JOB
    CASE
        WHEN JOB='SALESMAN'
        THEN '1'
        WHEN JOB = 'MANAGER'
        THEN '2'
        ELSE '3'
    END AS JOB_CODE
FROM emp
"""

_db.executeAll(sql)
```

```python
## concat --> 결합 sql에서도 CONCAT 

sql = """
select concat(ENAME, '(',JOB,')') as ENAME from emp
"""

_db.executeAll(sql)
```

```python
## 파생변수 생성

df["ENAME_"] = df["ENAME"] + '('+ df["JOB"] + ')'
df["ENAME_"]
```

```python
# if문을 사용한 쿼리문. if(조건문,참 값,거짓 값)
## 200이상의 값을 가지면 HIGH 아니면 LOW가 출력되는 파생변수 생성

sql = """
select SAL, if( SAL >= 200, "HIGH", "LOW") as "H/L" from emp 
"""

_db.executeAll(sql)
```

```python
df["H/L"] = df["SAL"].apply(lambda x: "HIGH" if x >=2000 else "LOW")
df[["SAL","H/L"]]
```

```python
def sal_change(x):
    if x > 2000:
        return "High"
    else:
        return "LOW"

df["H/L"]=df["SAL"].apply(sal_change)
df[["SAL","H/L"]]
```

```python
### pandas merge() 함수를 이용하여 left,right,inner 3가지의 경우를 모두 출력
### sql에서 했던 조건과 같다. [부서이름을 기준으로 merge 사용]
sql = """
select * from emp
"""

df= _db.executeAll(sql)
df
```

```python
sql = """
select * from dept
"""

df_2= _db.executeAll(sql)
df_2
```

Left Join & Right Join [판다스 pandas사용]

```python
## Left join

pd.merge(df, df_2, on="DEPTNO", how='left')
```

```python
## right join

pd.merge(df, df_2, on = "DEPTNO", how='right')
```

```python
## inner join

pd.merge(df, df_2, on = "DEPTNO", how = "inner")
```

```python
## full join

pd.merge(df,df_2, on = 'DEPTNO', how='right')
```

```python
## full outer join

pd.merge(df, df_2, on = 'DEPTNO', how='outer')
```
