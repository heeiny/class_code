```python
sql = """
select * from test
"""
## Database 클래스 안에서 결과값을 받아오는 함수 1개 -> executeAll()  결과값을 return을 해줌
## executeAll()  sql문을 실행, 결과값을 받아와서 그 값을 데이터프레임으로 변경해서 return

_db.executeAll(sql,values=[])

## _db.executeAll(sql)  매개변수 인자값이 일치해야 함.
```

```python
## Country기준으로 그룹화 Units Sold의 평균값, 합계

sql = """
select 
Country, 
AVG(`Units Sold`) as avg, 
COUNT(`Units Sold`) as cnt, 
SUM(`Units Sold`) as sum 
from sales 
group by Country
"""

_db.executeAll(sql)
```

```python
## 컬럼에서 중복 값을 제거하고 값들은 하나씩만 출력
sql = """
select 
distinct(Region) as region 
from sales
"""
_db.executeAll(sql)
```
