```python
## ruiage.csv파일 로드
import pandas as pd
uriage_data = pd.read_csv("../csv/uriage.csv")
uriage_data.head()
```

```python
## purchase_data칼럼의 데이터 타입 시계열로 변경

uriage_data["purchase_date"] = pd.to_datetime(uriage_data["purchase_date"])
```

```python
## item_name 칼럼의 값들을 대문자로 변환
uriage_data['item_name']=uriage_data['item_name'].str.upper()
uriage_data
# uriage_data["item_name"].apply(lambda x: x.upper())            ##위의 함수랑 같은 식
```

```python
uriage_data.head()
```

```python
## item_name 칼럼의 공백을 지운다. 
# strip -> 문자열 양 옆에 공백을 지우는 함수
# uriage_data["item_name"].str.replace(""," ")

uriage_data["item_name"]= uriage_data["item_name"].apply(lambda x: x.replace(" ",""))
```

```python
## item_name 칼럼을 기준으로 오름차순 

uriage_data.sort_values("item_name",ascending = True)
# uriage_data =uriage_data.sort_values("item_name",ascending = True)
```

```python
## 데이터의 결측치를 확인

uriage_data.isna().sum()
```

```python
uriage_data.isna().any(axis = 0 )           #False면 결측치 없음 , Ture면 결측치 존재
```

# item_price 결측치의 값을 해당하는 item_name들의 맥스값으로 치환하여 결측치를 채워주는 작업

```python
#1. flg_is_null이라고 임의로 이름을 지정해 줌

flg_is_null = uriage_data["item_price"].isna()
```

```python
#2. item_price의 값이 NaN인 item_name들의 중복 값을 삭제한 array형태  ##결측치가 있는 상품들의 list를 출력

uriage_data.loc[flg_is_null,"item_name"].unique()
```

```python
# ~연산자는 bool형태(참,거짓)의 데이터를 반전 시켜준다. 
## item의 name이 "상품 A"이면서 ~flg_is_null :아이템의 가격이 null이 아닌 가격 값들의 최댓값

uriage_data.loc[(~flg_is_null) & (uriage_data["item_name"] == "상품A"), "item_price"].max()
```

```python
uriage_data.loc[()]
```

```python
for trg in list(uriage_data.loc[flg_is_null,"item_name"].unique()):
    price = uriage_data.loc[(~flg_is_null)&(uriage_data["item_name"] == trg),    ## price= 가격이 null이 아닌 값& 결측치를 제거한 item name중에서 
                            "item_price"].max()                                    ##가격이 최댓값
    # print(trg,price)
   
    uriage_data["item_price"].loc[(flg_is_null) & (uriage_data ["item_name"] == trg)] = price

    ##각각 item에 맞는 결측치에 max값을 넣어줌
```
