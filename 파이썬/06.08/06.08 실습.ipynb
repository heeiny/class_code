```python
## reset이 되기전의 인덱스의 값을 초기화 하는 함수(넘버링을 0부터 함) ->매개변수 drop
## drop이라는 속성은 기존에 가지고 있던 인덱스의 값을 index컬럼에 유지를 할지를 정하는 속성(True이면 유지)
## 만약 Falsf값이 나오면 새로운 column이 생성됨

df_3.reset_index(drop=True, inplace=True)  
df_3
```

```python
## inplace속성
df_3 = df_3.sort_values("권역")
#df_3에 df_3 데이터프레임에 '권역'컬럼을 기준으로 오름차순 정렬을 한 데이터프레임을 할당
#이 코드를 실행을 하면 df_3라는 데이터프레임이 변경

df_3.sort_value("권역",inplace=True)
#df_3 = df_3.sort_values("권역")과 같은 의미
```

```python
df_3["총이윤"] = df_3["총수익"] - df_3["총 비용"]  ##새로운 column생성
df_3.head()
```

```python
## df_3 데이터프레임에서 총수익,총비용,총이윤이라는 파생변수를 생성.
## 주가,부실기업 등 기존 데이터를 이용하여 예측의 정밀도를 높임
```

```python
## 총수익 -> 판매단위 * 단가
## 총비용 -> 판매단위 * 원가
## 총이윤 -> 총수익 - 총비용
```

```python
df_3["총수익"] = df_3["판매단위"] * df_3["단가"]
df_3["총비용"] = df_3["판매단위"] * df_3["원가"]
df_3["총이윤"] = df_3["총수익"] - df_3["총비용"]
```

```python
df_3.head()
```

```python
df_3.drop(["총수익","총비용","총이윤"], axis=1, inplace=True)               ##inplace=True"의 의미는, 해당 변화를 즉각적으로 해당 DataFrame에 반영하겠다는 의미입니다. 
                                                                            ##False로 한다면 새로운 DataFrame에 변경 내용을 담아주도록 해야 합니다.
```
